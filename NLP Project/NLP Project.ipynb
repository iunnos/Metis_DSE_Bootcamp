```python
import re
stemmer = nltk.PorterStemmer()
def prep(word):
    word=word.lower() 
    word=re.sub("\\W"," ",word) # remove special chars
    word=re.sub("\\s+(in|the|all|for|and|on)\\s+"," _connector_ ",word) # normalize certain words
    word=re.sub("\\d",'',word) # remove numbers
    
    # stem words
    words=re.split("\s+",word)
    stemmed_words=[stemmer.stem(word=word) for word in words]
    return ' '.join(stemmed_words)
```

```python
#Thanks to Braden Taack for his functions
def top_reviews(topic_df, topic, n_reviews):
    return (topic_df
            .sort_values(by=topic, ascending=False)
            .head(n_reviews)['raw_reviews']
            .values)

def top_words(word_topic_df, topic, n_words):
    return (word_topic_df
            .sort_values(by=topic, ascending=False)
            .head(n_words))[topic]

def analyze_topic(topic_df, word_topic_df, topic='topic_0', n_review=10,n_words=10 ):
    for review in top_reviews(topic_df, topic, n_review):
        print(review)
        print()
    print('The Top Words for this Topic Are: ')
    print(top_words(word_topic_df, topic, n_words))
```

## Recommender
- Goal: Recommend similar other yelp users that provide similar reviews
- What I have to work with:
    1. NMF Matrix (dist to similar reviews)
    2. Business location (provide only users that are in the specified location)
- Things to keep in mind:
    1. Dataframe (topic_df) rows represents a single review. A user can/usually has multiple reviews. user_ids will get repeated
    2. NMF matrix provides distance similarity to other reviews. ie. looking at the similarity of different reviews.

```python
cache = dict()
def user_rec (userID,nmf_matrix=nmf_matrix,topic_df=topic_df):
    if userID in cache:
        return cache[userID].keys()
    users=[]
    for l in topic_df.loc[topic_df['user_id']==userID].index:
        for j in range(200): #(nmf_matrix.shape[0]): more so as a proof of concept. range reduced to 200 instead of complete range of matrix
            if topic_df.iloc[j].user_id != userID:
                sim = np.dot(nmf_matrix[j,:],nmf_matrix[l,:])
                users.append((topic_df.iloc[j].user_id,sim))

    users.sort(key=lambda x: x[1], reverse=True)
    seen=set()
    top_users = [(a, b) for a, b in users if not (a in seen or seen.add(a))]
    cache[userID]=top_users
    return [top[0] for top in top_users[0:5]]
```
